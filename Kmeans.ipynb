```python
import pandas as pd
```

```python
data = pd.read_csv('Kmeans_happyscore_income.csv')
data.head()
```

```python
import matplotlib.pyplot as plt

plt.xlabel('Happy Score')
plt.ylabel('Avg Income')
plt.scatter(data['happyScore'], data['avg_income'])
```

```python
data.sort_values('avg_income', inplace = True) #inplace True means data is arranged and saved that way
```

```python
richest = data[data.avg_income > 15000]
```

```python

plt.xlabel('Happy Score')
plt.ylabel('Avg Income')
plt.scatter(richest['avg_income'], richest['happyScore'])

# plt.text(x_axis,y_axis,text_lable_required)
plt.text(richest.iloc[0]['avg_income'],
         richest.iloc[0]['happyScore'],
         richest.iloc[0]['country'])

plt.text(richest.iloc[-1]['avg_income'],
         richest.iloc[-1]['happyScore'],
         richest.iloc[-1]['country'])
```

```python
plt.scatter(richest['avg_income'], richest['happyScore'])

for k, row in richest.iterrows():
    plt.text(row['avg_income'],
             row['happyScore'],
             row['country'])
```

```python
plt.scatter(data['avg_income'], data['happyScore'], s = data['income_inequality']*10, alpha=0.25)
```

```python
from sklearn.cluster import KMeans
import numpy as np

income = data['avg_income']
happy = data['happyScore']
ineq = data['income_inequality']

income_happy = np.column_stack((income, happy))
```

```python
km_res = KMeans(n_clusters=3).fit(income_happy)
clusters = km_res.cluster_centers_
```

```python
plt.scatter(income, happy)
plt.scatter(clusters[:,0], clusters[:,1], s=500)
```

```python
data = pd.read_csv('Banknote-authentication-dataset.csv')
data.head()
```

```python
from sklearn.cluster import KMeans
import numpy as np

v1 = data['V1']
v2 = data['V2']

combined = np.column_stack((v1, v2))

km_res = KMeans(n_clusters=3).fit(combined)
clusters = km_res.cluster_centers_

plt.scatter(v1, v2)
plt.scatter(clusters[:,0], clusters[:,1], s=500)
```
